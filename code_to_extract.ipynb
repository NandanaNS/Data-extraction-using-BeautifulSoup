### Function for extracting:

### Table1:

```python
def jsondicttable1(schoolrow):
    list1=[]
    for i in schoolrow: 
        #key1:
        col0=i.find_all('td')[0].text.replace('\n','').strip()
        #value1:
        col1=i.find_all('td')[2].text.replace('\n','').strip()
        #key2 
        col2=i.find_all('td')[3].text.replace('\n','').strip()
        #value2:
        col3=i.find_all('td')[5].text.replace('\n','').strip()
        #key3:
        col4=i.find_all('td')[6].text.replace('\n','').strip()
        #value3:
        col5=i.find_all('td')[8].text.replace('\n','').strip()
        
        # Create a dictionary with the above information
        data = {col0: col1,col2: col3,col4: col5}
        
        # Append the dictionary to the list
        list1.append(data)
    d={}
    for dictionary in list1:
        d.update(dictionary)
    return d
```

### Table2:

```python
def jsondicttable2(schoolrow):
    list1=[]
    for i in schoolrow: 
        #key1:
        col0=i.find_all('td')[0].text.replace('\n','').strip()
        col0=col0.split(".",1)[1]
        #value1:
        col1=i.find_all('td')[2].text.replace('\n','').strip()
        #key2 
        col2=i.find_all('td')[3].text.replace('\n','').strip()
        col2=col2.split(".",1)[1]
        #value2:
        col3=i.find_all('td')[5].text.replace('\n','').strip()
        #key3:
        col4=i.find_all('td')[6].text.replace('\n','').strip()
        col4=col4.split(".",1)[1]
        #value3:
        col5=i.find_all('td')[8].text.replace('\n','').strip()
        
        # Create a dictionary with the above information
        data = {col0: col1,col2: col3,col4: col5}
        
        # Append the dictionary to the list
        list1.append(data)
    d={}
    for dictionary in list1:
        d.update(dictionary)
    return d
```

### Table3:

```python
def jsondicttable3(schoolrow):
    table3_list=[]
    for i in schoolrow:
        schoolname=i.text
        desig=schoolname.rsplit("-",1)[0]
        num=schoolname.rsplit("-",1)[1]
        data = {desig: num}
        table3_list.append(data)
    d={}
    for dictionary in table3_list:
        d.update(dictionary)
    return d
```

### Table 4 :

```python
def jsondicttable4(schoolrow2):
    list45=[]
    for i in schoolrow2: 
        list45.append(i.text.replace('\n','  ').strip()) #removing \n from rows
    list45.pop(0) #removing 'standard'
    my_array = [phrase.split() for phrase in list45]  # splitting each rows into seperate list of strings
    listind45=[]
    for j in range(0,len(my_array[0])):
        for k in range(0,3):
            line1=my_array[0][j]+'_'+my_array[1][k] #'Malayalam_Boys'
            listind45.append(line1)
    my_array.pop(0)  #removing ['Malayalam', 'English', 'Tamil', 'Kannada', 'ALL']
    my_array.pop(0)  #remove Boys','Girls','Total
    my_array.pop(-1)  #removing 'total' row
    ind2=[]
    for w in range(0,len(my_array)):
        ind2.append(my_array[w][0]) # ind2=['8', '9', '10']
    #creating the keys:
    indfinal=[]
    for b in range(0,len(ind2)):
      for a in range(0,len(listind45)):
            line1=listind45[a]+'_'+ind2[b]  #Malayalam_Boys_8'
            indfinal.append(line1)
    for c in range(0,len(my_array)):
        del my_array[c][0]   #removing ['8', '9', '10']
    #getting the values:
    val1=[]
    for d in range(0,len(my_array)):
        for e in range(0,len(my_array[d])):
            val1.append(my_array[d][e])  #making all values into continuos list
    #making key value pair:
    mylist4=[]
    for f in range(0,len(indfinal)):
        key=indfinal[f]
        value=val1[f]
        data={key:value}
        mylist4.append(data)  # assigning first value of index list to first value ofvalue list.....
    #converting the list to dictionary
    d={}
    for dictionary in mylist4:
        d.update(dictionary)
    return d
```

## Table5:

```python
def jsondicttable5(schoolrow2):
    list45=[]
    for i in schoolrow2: 
        list45.append(i.text.replace('\n','  ').strip())
    list45.pop(0)
    my_array = [phrase.split() for phrase in list45]
    listind45=[]
    for j in range(0,len(my_array[0])):
        for k in range(0,3):
            line1=my_array[0][j]+'_'+my_array[1][k]
            listind45.append(line1)
    my_array.pop(0)
    my_array.pop(0)
    ind2=[]
    for w in range(0,len(my_array)):
        ind2.append(my_array[w][0]+my_array[w][1])
    #creating the keys:
    indfinal=[]
    for b in range(0,len(ind2)):
      for a in range(0,len(listind45)):
            line1=listind45[a]+'_'+ind2[b]
            indfinal.append(line1)
    for c in range(0,len(my_array)):
        del my_array[c][0]
        del my_array[c][0]
    #getting the values:
    val1=[]
    for d in range(0,len(my_array)):
        for e in range(0,len(my_array[d])):
            val1.append(my_array[d][e])
    #making key value pair:
    mylist4=[]
    for f in range(0,len(indfinal)):
        key=indfinal[f]
        try: 
            value=val1[f]
        except IndexError:
            value='0'
        data={key:value} 
        mylist4.append(data)
    #converting the list to dictionary
    d={}
    for dictionary in mylist4:
        d.update(dictionary)
    return d
```

## Table6

```python
def jsondicttable6(schooltd):
    list45=[]
    for i in schooltd: 
        list45.append(i.text.replace('\n','  ').strip())
    middle_index=len(list45)//2
    keylist=list45[:middle_index]
    valuelist=list45[middle_index:]
    listfor6=[]
    for i in range(0,middle_index):
        key='vhss_'+keylist[i]
        value=valuelist[i]
        data={key:value}
        listfor6.append(data)
    d={}
    for dictionary in listfor6:
        d.update(dictionary)
    return d
```

## Combining all the tables:

```python
def scl_tables(i):
    #mylist=[]
    html_text=requests.get(i).text
    soup=BeautifulSoup(html_text,'lxml') 
    
    #table1:
    scl1=soup.find_all('tbody')[0]  # our table is inside the tbody tag.
    schoolrow=scl1.find_all('tr')
    my_json=jsondicttable1(schoolrow) 
    
    #table2:
    scl2=soup.find_all('tbody')[1]  # our table is inside the tbody tag.
    schoolrow2=scl2.find_all('tr')
    my_json2=jsondicttable2(schoolrow2)
    
    #table3:
    scl3=soup.find_all('tbody')[2] # our table is inside the tbody tag.
    schoolrow3=scl3.find_all('a')
    my_json3=jsondicttable3(schoolrow3)
    
    #table4 :
    div4 = soup.find('div', attrs={'id':'students-lpuphs'})
    scl4=div4.find('tbody') # our table is inside the tbody tag.    
    schoolrow4=scl4.find_all('tr')
    my_json4=jsondicttable4(schoolrow4)
    
    #table5:
    div5 = soup.find('div', attrs={'id':'students-hss'})
    try:
        scl5=div5.find('tbody') # our table is inside the tbody tag.   
        schoolrow5=scl5.find_all('tr')
        my_json5=jsondicttable5(schoolrow5)
        onescldict={**my_json,**my_json2,**my_json3,**my_json4,**my_json5}
    except AttributeError:
        onescldict={**my_json,**my_json2,**my_json3,**my_json4}
    
    #table6:
    div6 = soup.find('div', attrs={'id':'students-vhss'})
    try: 
        scl6=div6.find('tbody')  # our table is inside the tbody tag. 
        schoolrow6=scl6.find_all('tr') 
        schooltd=scl6.find_all('td')
        scl6=jsondicttable6(schooltd)
        onescldict.update(scl6)
    except:
        exit
    #print(onescldict)
    return onescldict
```

### Driver code:

```python
for i in l2:
    filename=i.rsplit("/",1)[1]
    scls=scl_tables(i)
    scljson=json.dumps(scls)
    with open(filename+'.json','w') as f:
        f.write(scljson)
        f.close()
```
