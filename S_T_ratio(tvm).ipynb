```python
import os
import re
import json
import pandas as pd
```

```python
folderPath='/content/tvm/'
os.chdir(folderPath)
data = []
    
    #Iterating through the files in the folder and appending all the data to the list
for filename in os.listdir():
    if filename.endswith('.json'):
        with open(filename) as f:
            d = json.load(f)
            dct = {k:[v] for k,v in d.items()}
            df = pd.DataFrame(dct)
            data.append(df)
final_data = pd.concat(data)
```

```python
mydata=final_data
mydata.shape
```

## creating a dataframe of employee details alone:

```python
mydata_a=mydata.iloc[:,75:]      #dataframe with columns starting from 75th .
all_=mydata_a.columns.values.tolist()       #taking a list of the above created dataframe.
st=mydata_a.filter(like="_")         #dataframe with columns having _ in the column name.
st=st.columns.values.tolist()        #list of columns in this dataframe.

#removing all columns with _ fromthe table mydata_a
faculty=[]
for i in all_ :
    if i not in st:
        faculty.append(i)
faculty_only=mydata.loc[:,faculty]
droplist=[' Total Employees ','LP Section Phone','LP Section Email','Section Phone','Section Email','UP Section Phone','UP Section Email']
for j in droplist:
    if j in faculty_only.columns:
        faculty_only.drop(j,axis=1,inplace = True)  
```

```python
faculty_only.shape
```

```python
faculty_count=faculty_only.count().to_frame(name='faculty_count')
```

```python
faculty_count.T
```

```python
# list to keep track of classified columns:
classified=[]
def classified_columns(li):
    for a in li:
        classified.append(a)
```
